No seguinte projeto farei uma análise dos dados de covid-19 nos últimos anos e também farei uma análise em uma determinada data, para trazer valores da pandemia em diferentes momentos, e trarei algumas situações que considero cabiveis dentro desses dados, possibilitando a análise e exploração dos dados, com o estudo focado no Brasil, mas com a possibilidade de se trabalhar com outros paises, caso se queira.

```python
import pandas as pd

#os links referentes aos dados utilizados(CSSEGISandData/Covid-19)
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
confirmed = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
```

Aqui vemos que os dados são bem semelhantes em sua estrutura, entretanto fica perceptivel que temos uma coluna com valores NA, então já vamos retirar essa coluna.

```python
confirmed_by_country = confirmed.groupby("Country/Region").sum().dropna()
#colocando os dados agrupados por pais para a analise e removenedo os valores 'na'
```

## Unindo os dados
Agora vou agrupar todos os dados dos 3 dataframes. Aqui, vou definir uma data para a uniao e para termos uma analise mais focada em um periodo, mas a data pode ser alterada facilmente, onde inicialmente vou escolher a ultima data disponivel (importante observar o formato em que a data é representada dentro do nossos dados, caso contrário teremos um erro nos dados).

```python
last_date = "3/9/23"
confirmed[last_date]
```

Aqui é necessária muita atenção, pois se o pais ou o dataset for escrito de forma errada o codigo lançara um erro. Pode-se fazer uma conferencia da lista de valores unicos dentro desse conjunto de dados, encontrando todos os nomes corretos para cada região antes de ativar a função. O código abaixo é para verificar todos os valores.

```python
#confirmed['Country/Region'].unique()
```

```python
def lastest_by_country(data, pais):

  assert pais in data['Country/Region'].unique(), f"{pais} não está na lista de Country/Region."
  
  filtro = data[data['Country/Region'] == pais]

  if data.empty:
    raise ValueError("'Não tem-se dados com o nome fornecido'")
  
  return filtro.groupby('Country/Region').sum()[last_date].head()
#codigo para pegar as informaçoes dos conjuntos de dados
```

```python
lastest_by_country(confirmed, 'Brazil')
```

```python
informations = [lastest_by_country(confirmed, 'Brazil'), lastest_by_country(deaths, 'Brazil'), lastest_by_country(recovered, 'Brazil')]
informations_df = pd.concat(informations,  axis=1)
informations_df.columns = ['Confirmados','Mortes', 'Recuperados' ]
#o numero de recuperados na data proposta, data final do conjunto de dados, é nula devido ao avanço da imunização
#caso ocorra uma mudança de data os valores irao ser alterados
informations_df
```

## Taxa de letalidade Pós pandemia
Agora, temos o nosso dataFrame especialmente trabalhado para os dados do Brasil e podemos calcular a taxa de letalidade sem grande problemas, uma vez que essa taxa na verdade é de quantas pessoas faleceram da doença na época avaliada, dado em porcentagem.

```python
letalidade = round((informations_df.Mortes / (informations_df.Confirmados+informations_df.Recuperados))*100,2)
letalidade
```

Podemos observar então uma mortalidade de 1,89% de morte. Isso ocorre devido ao fato da análise se fazer depois dos piores momentos da pandemias e da larga imunização da população com as VACINAS, o que foi crucial para diminuir a taxa de mortalidade dos pacientes aqui no Brasil. 

A mesma coisa pode ser feita para outros paises do nosso conjunto de dados, assim como em outra data para saber a situação da pandemia em determinada data, que será feita logo em sequência.

## Análise em diferente data

Aqui eu farei uma análise em um outro dia qualquer, em 2021 por exemplo, onde tinhamos uma maior quantidade de pessoas afetadas pela doença, mas ainda não tinhamos a vacina para auxiliar no controle de letalidade do Covid-19.

```python
last_date = "10/10/21"

informations = [lastest_by_country(confirmed, 'Brazil'), lastest_by_country(deaths, 'Brazil'), lastest_by_country(recovered, 'Brazil')]
informations_df = pd.concat(informations,  axis=1)
informations_df.columns = ['Confirmados','Mortes', 'Recuperados' ]


letalidade = round((informations_df.Mortes / (informations_df.Confirmados+informations_df.Recuperados))*100,2)
letalidade
```

Aqui ja temos um aumento da mortalidade do Covid, chegando a 2,79%, devido ao dia análisado.

# Comparação entre os paises
Agora, vou comparar esses dados do Brasil com o dado de outros paises. Importante lembrar que estamos nos baseando na data '10/10/21', em outras datas teremos valores diferentes de forma global, entao é importante a atenção a esse fato antes de uma interpretação mais drásticas desses dados.

```python
plt.figure(figsize=(10,5))

confirmed_by_country[last_date].sort_values(ascending=False)[1:10].plot(kind='bar')
plt.title(f'Confirmados por pais (10 primeiros paises em {last_date})')

plt.show()
```

Destacando o Brasil do restante do mundo nos gráficos podemos fazer algumas análises para essa data em especifico.

```python
eh_Brasil = confirmed['Country/Region'] == 'Brazil'
confirmed['eh_Brasil'] = eh_Brasil

confirmed['eh_Brasil'] = confirmed['eh_Brasil'].replace({False: 'outro_pais', True: 'Brasil'})
#alterando os valores booleans
#confirmed.head()
```

```python
confirmed.head(2)
```

Agora que adicionamos uma coluna para verificar se a regiao da linha em si é a do Brasil ou não, podemos trabalhar com essas coluna de forma que fica possivel a comparação com outros paises.

```python
soma_de_casos_confirmados = confirmed.groupby('eh_Brasil').sum()
```

```python
confirmed.groupby('eh_Brasil').sum()[last_date].plot(kind='bar')
labels = ['Brasil', 'Não é Brasil']
plt.xticks([0, 1], labels)
plt.title(f'Comparação de casos confirmados no mundo e no Brasil até {last_date}');
```

```python
soma_de_casos_confirmados.T.plot();
#O T permite a inversão para transformar as linhas em colunas e as colunas em linhas.
```

Fica nitido como o crescimento de uma pandemia ocorre de forma acentuada de forma global e também quando analisado em um único país.

```python
diferencas_dia_a_dia = soma_de_casos_confirmados.T[2:-2].diff().dropna()
diferencas_dia_a_dia.plot()
plt.title('Casos novos durante os dias');
```

# Trabalhando com os casos
Agora, também podemos trabalhar melhor alguns valores, como entender a incidência, eventos que ocorrem novamente dentro de um espaço de tempo dividido pela população que é afetada pelo o que estamos estudando, no caso seria a taxa de contaminação pela população nacional no espaço de tempo analisado.

```python
casos = confirmed.groupby('eh_Brasil').sum()[last_date]['Brasil']
casos
```

Tendo em mente que a populção do Brasil chegou próxima a 203 milhões de habitantes em 2023 (esse valor sofre alteração de ano a ano, entao devemos ter atenção com o valor da população combinada com o last date) podemos aplicar a idéia de incidência.

```python
populacao_Brasil = 203000000

incidencia = round((casos/populacao_Brasil)*100,2)
print(f'A taxa de incidência do covid é de {incidencia}%')
```

<b>Taxa de incidência:</b> Qual a chance de alguém adquirir a doença nesse periodo analisado para essa população.

A taxa de incidência leva em conta também quantas pessoas foram confirmadas com a doença, o que no caso pode ter grandes alterações de valores corretos devido ao incentivo governamental de cada país para encontrar os infectados e registra-los, por exemplo na Coreia do Sul o governo incentivou a população a ir fazer os testes, logo com mais pessoas testadas foram maiores a taxa de incidencia no pais.

```python
eh_Brasil = deaths['Country/Region'] == 'Brazil'
deaths['eh_Brasil'] = eh_Brasil

deaths['eh_Brasil'] = confirmed['eh_Brasil'].replace({False: 'outro_pais', True: 'Brasil'})
```

```python
mortes = deaths.groupby('eh_Brasil').sum()[last_date]['Brasil']

taxa_de_mortes = round((mortes/populacao_Brasil)*100,3)
print(f'A taxa de mortalidade do covid é de {taxa_de_mortes}%')
```

## Gráficos de comparação final

```python
agrupados_por_dia = pd.DataFrame([confirmed.sum()[2:-1], recovered.sum()[2:], deaths.sum()[2:]],
                                 index=['confirmados', 'recuperados','mortos']).T
agrupados_por_dia = agrupados_por_dia.drop(agrupados_por_dia.index[0]) #removendo a longitude
agrupados_por_dia = agrupados_por_dia.drop(agrupados_por_dia.index[-1]) #removendo a longitude
agrupados_por_dia.head()
```

Adicionando a taxa de letalidade ao dataframe agora

```python
agrupados_por_dia['taxa_de_letalidade'] = (agrupados_por_dia['mortos']/(agrupados_por_dia['mortos']+agrupados_por_dia['recuperados']))*100
agrupados_por_dia
```

```python
agrupados_por_dia['taxa_de_letalidade'].plot()
plt.title('Taxa de letalidade por toda a pandemia em porcentagem')
```

Como fica visivel nosso dados param de resgistrar os recuperados apatir de uma determinada data, ocorrendo o salto no gráfico para 1, uma vez que fica dividido os valores de mortes pelo proprio valores de mortes, entao vamos remover esses valores e refazer nosso gráfico.

```python
teste = agrupados_por_dia.query('taxa_de_letalidade==100')
teste.head(1)
```

```python
agrupados_por_dia.index = pd.to_datetime(agrupados_por_dia.index)

data_filtrada = agrupados_por_dia.loc[agrupados_por_dia.index < '2021-08-05']
# Filtrando as linhas onde a data é anterior à '8/5/21'

plt.figure(figsize=(10,5))
plt.title("Taxa de letalidade no decorrer do tempo no Brasil, entre 22/01/2020 e 05/08/2021, em porcentagem.")

data_filtrada['taxa_de_letalidade'].plot()
# Plotando os dados filtrados
plt.show()
```
